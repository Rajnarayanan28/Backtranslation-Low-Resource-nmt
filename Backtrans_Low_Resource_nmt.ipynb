```python
import random

# Malayalam synonym dictionary
ml_synonyms = {
    "വലിയ": ["വിപുലം", "മഹത്തായത്"],
    "നല്ല": ["ഉത്തമം", "ശ്രേഷ്ഠം"],
    "ചെറിയ": ["ലഘു", "ചുരുങ്ങിയ"],
    "കുട്ടി": ["ശിശു", "ബാലിക"],
    "വീട്": ["ഗൃഹം", "നിലയം"]
}

# Synonym replacement
def synonym_replace(text):
    words = text.split()
    new_words = []
    for w in words:
        if w in ml_synonyms and random.random() < 0.3:
            new_words.append(random.choice(ml_synonyms[w]))
        else:
            new_words.append(w)
    return " ".join(new_words)

# Safe word shuffle
def shuffle_sentence(text):
    words = text.split()
    if len(words) > 5:
        mid = words[1:-1]
        random.shuffle(mid)
        return " ".join([words[0]] + mid + [words[-1]])
    return text

# Tense variation
def tense_shift(text):
    text = text.replace("ന്നു", "ും")
    text = text.replace("ിച്ചു", "ിക്കും")
    return text

# Dative -> nominative
def dative_to_nominative(text):
    if "എനിക്ക്" in text:
        return text.replace("എനിക്ക്", "ഞാൻ").replace("തോന്നി", "വിചാരിച്ചു")
    if "അവന്" in text:
        return text.replace("അവന്", "അവൻ")
    return text

# Passive -> active
def passive_to_active(text):
    if "ചെയ്തു" in text:
        return text.replace("ചെയ്തു", "ചെയ്യുന്നു")
    return text
```

```python
aug_ml = []
aug_en = []

for i, row in combined_df.iterrows():
    ml = row["ml"]
    en = row["en"]

    variants = [
        synonym_replace(ml),
        shuffle_sentence(ml),
        tense_shift(ml),
        dative_to_nominative(ml),
        passive_to_active(ml)
    ]

    for v in variants:
        if v != ml:
            aug_ml.append(v)
            aug_en.append(en)
```

```python
aug_df = pd.DataFrame({
    "ml": aug_ml,
    "en": aug_en
})

print("Augmentation size:", len(aug_df))
aug_df.head()
```

```python
final_df = pd.concat([combined_df, aug_df], ignore_index=True)
final_df.to_csv("/content/final_augmented_dataset.csv", index=False, encoding="utf-8")

print("Final dataset size:", len(final_df))
```

```python
# CONFIG
model_name = "Helsinki-NLP/opus-mt-ml-en"
csv_path = "/content/final_augmented_dataset.csv"
output_dir = "/content/final_mt_model"

# LOAD DATA
df = pd.read_csv(csv_path)
df = df.dropna().reset_index(drop=True)
print("Loaded rows:", len(df))

dataset = Dataset.from_pandas(df)
dataset = dataset.train_test_split(test_size=0.10, seed=42)

train_data = dataset["train"]
eval_data = dataset["test"]

# LOAD TOKENIZER & MODEL
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# TOKENIZATION
max_source_length = 128
max_target_length = 128

def preprocess(batch):
    model_inputs = tokenizer(
        batch["ml"],
        max_length=max_source_length,
        truncation=True,
        padding="max_length"
    )
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            batch["en"],
            max_length=max_target_length,
            truncation=True,
            padding="max_length"
        )

    labels["input_ids"] = [
        [(token if token != tokenizer.pad_token_id else -100) for token in seq]
        for seq in labels["input_ids"]
    ]

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

train_dataset = train_data.map(preprocess, batched=True, remove_columns=train_data.column_names)
eval_dataset = eval_data.map(preprocess, batched=True, remove_columns=eval_data.column_names)

#  DATA COLLATOR
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)

# BLEU METRIC (Updated Using evaluate)
bleu_metric = evaluate.load("sacrebleu")

def postprocess(preds, labels):
    preds = [p.strip() for p in preds]
    labels = [[l.strip()] for l in labels]
    return preds, labels

def compute_metrics(eval_pred):
    preds, labels = eval_pred
    if isinstance(preds, tuple):
        preds = preds[0]

    decoded_preds = tokenizer.batch_decode(preds, skip_special_tokens=True)

    labels = [
        [(l if l != -100 else tokenizer.pad_token_id) for l in label]
        for label in labels
    ]
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    decoded_preds, decoded_labels = postprocess(decoded_preds, decoded_labels)
    bleu = bleu_metric.compute(predictions=decoded_preds, references=decoded_labels)

    return {"bleu": bleu["score"]}

# TRAINING ARGUMENTS
training_args = Seq2SeqTrainingArguments(
    output_dir=output_dir,
    do_train=True,
    do_eval=True,
    eval_strategy="epoch",
    learning_rate=3e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=4,
    weight_decay=0.01,
    save_total_limit=3,
    save_strategy="epoch",
    predict_with_generate=True,
    logging_steps=200,
    fp16=torch.cuda.is_available(),
    report_to="none",
    load_best_model_at_end=True,
    metric_for_best_model="bleu",
    greater_is_better=True
)


# TRAINER
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

# TRAIN
trainer.train()

trainer.save_model(output_dir)
tokenizer.save_pretrained(output_dir)

print("✔ Training finished. Model saved to:", output_dir)
```
